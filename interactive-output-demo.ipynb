# Interactive Output Demo

This notebook exercises the new rich output pipeline in Nebula.

It includes:
- standard Jupyter `application/vnd.plotly.v1+json`
- Nebula-native `application/vnd.nebula.web+json`
- a Nebula widget that reuses the shared Plotly runtime
- isolated HTML output via Jupyter metadata

Run the code cells top to bottom.

```python
from IPython.display import display

plotly_payload = {
    "data": [
        {
            "type": "bar",
            "x": ["Alpha", "Beta", "Gamma", "Delta"],
            "y": [5, 8, 3, 9],
            "marker": {"color": ["#0f766e", "#2563eb", "#f59e0b", "#dc2626"]},
        }
    ],
    "layout": {
        "title": "Plotly MIME Demo",
        "height": 360,
        "margin": {"l": 40, "r": 20, "t": 60, "b": 40},
        "paper_bgcolor": "white",
        "plot_bgcolor": "white",
    },
    "config": {"displayModeBar": False, "responsive": True},
}

display(
    {
        "application/vnd.plotly.v1+json": plotly_payload,
        "text/plain": "Plotly MIME demo",
    },
    raw=True,
)
```

```python
from IPython.display import display

nebula_widget = {
    "height": 260,
    "html": """
    <div data-nebula-root class='card'>
      <div class='eyebrow'>Nebula JS Output</div>
      <h2>Interactive Mini Widget</h2>
      <p class='subtitle'>This is running through <code>application/vnd.nebula.web+json</code>.</p>
      <div class='metric-row'>
        <span>Current value</span>
        <strong data-value>0</strong>
      </div>
      <button data-randomize>Shuffle values</button>
      <div data-bars class='bars'></div>
    </div>
    """,
    "css": """
    .card {
      font-family: ui-sans-serif, system-ui, sans-serif;
      padding: 18px;
      border-radius: 16px;
      color: #0f172a;
      background: linear-gradient(135deg, #f8fafc, #e0f2fe);
      border: 1px solid #cbd5e1;
      box-shadow: 0 8px 20px rgba(15, 23, 42, 0.08);
    }
    .eyebrow {
      font-size: 11px;
      letter-spacing: 0.12em;
      text-transform: uppercase;
      color: #0369a1;
      margin-bottom: 10px;
      font-weight: 700;
    }
    h2 {
      margin: 0 0 8px;
      font-size: 24px;
      line-height: 1.2;
    }
    .subtitle {
      margin: 0 0 16px;
      color: #334155;
      font-size: 14px;
    }
    code {
      background: rgba(255,255,255,0.7);
      padding: 2px 6px;
      border-radius: 999px;
      font-size: 12px;
    }
    .metric-row {
      display: flex;
      justify-content: space-between;
      align-items: center;
      margin-bottom: 14px;
      font-size: 14px;
    }
    strong {
      font-size: 28px;
      color: #7c3aed;
    }
    button {
      border: 0;
      background: #0f172a;
      color: white;
      padding: 10px 14px;
      border-radius: 999px;
      cursor: pointer;
      font-weight: 600;
      margin-bottom: 14px;
    }
    .bars {
      display: grid;
      grid-template-columns: repeat(5, minmax(0, 1fr));
      gap: 10px;
      align-items: end;
      height: 96px;
    }
    .bar {
      border-radius: 10px 10px 4px 4px;
      background: linear-gradient(180deg, #7c3aed, #2563eb);
      min-height: 16px;
      transition: height 160ms ease;
    }
    """,
    "js": """
    const state = [18, 34, 26, 42, 30];
    const barsEl = container.querySelector('[data-bars]');
    const valueEl = container.querySelector('[data-value]');
    const button = container.querySelector('[data-randomize]');

    const render = () => {
      valueEl.textContent = String(state.reduce((sum, value) => sum + value, 0));
      barsEl.replaceChildren(...state.map((value) => {
        const bar = document.createElement('div');
        bar.className = 'bar';
        bar.style.height = `${value * 2}px`;
        return bar;
      }));
    };

    button.addEventListener('click', () => {
      for (let i = 0; i < state.length; i += 1) {
        state[i] = 16 + Math.floor(Math.random() * 32);
      }
      render();
    });

    render();
    """,
}

display(
    {
        "application/vnd.nebula.web+json": nebula_widget,
        "text/plain": "Nebula widget demo",
    },
    raw=True,
)
```

```python
from IPython.display import display

nebula_plotly = {
    "height": 360,
    "html": "<div data-nebula-root style='width: 100%; height: 320px;'></div>",
    "libraries": ["plotly"],
    "data": {
        "x": ["Q1", "Q2", "Q3", "Q4"],
        "y": [12, 18, 15, 24],
    },
    "js": """
    await libraries.plotly.newPlot(
      container,
      [{
        type: 'scatter',
        mode: 'lines+markers',
        x: data.x,
        y: data.y,
        line: { color: '#0f766e', width: 3 },
        marker: { size: 10, color: '#f97316' },
      }],
      {
        title: 'Nebula Web + shared Plotly runtime',
        margin: { l: 40, r: 20, t: 54, b: 40 },
        paper_bgcolor: 'white',
        plot_bgcolor: 'white',
      },
      { displayModeBar: false, responsive: true },
    );

    return () => libraries.plotly.purge?.(container);
    """,
}

display(
    {
        "application/vnd.nebula.web+json": nebula_plotly,
        "text/plain": "Nebula Plotly demo",
    },
    raw=True,
)
```
